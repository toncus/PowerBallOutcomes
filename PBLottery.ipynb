# PowerBall Lottery Outcomes

### The purpose of this application is to get the top White Ball and PowerBall number outcomes from over 100 PowerBall drawings and their related perentages. These records were downloaded form NJ PowerBall website for dates 03/04/2017 through 03/07/2018. Some changes were made to the actual file for data readbility.

```python
# Remove unecessary columns from the DataFrame and save the new DataFrame
reduced_df = lottery_df[["Date Winning Numbers Power Play Annuity Payout"]]

reduced_df.head(45)
```

```python
drawingdate=reduced_df["Date Winning Numbers Power Play Annuity Payout"].str[0:10]
drawingdate=drawingdate.replace("\n"," ")
five_numbers=reduced_df["Date Winning Numbers Power Play Annuity Payout"].str[11:25]
five_numbers=five_numbers.replace("\n"," ")
power_ball =reduced_df["Date Winning Numbers Power Play Annuity Payout"].str[26:28]
power_ball=power_ball.replace("\n"," ")
```

```python
# Builld DataFrame
Lottery_newdf = pd.DataFrame({"Date": drawingdate,
                          "Winning Numbers": five_numbers,
                          "Power": power_ball})
Lottery_newdf=Lottery_newdf[["Date","Winning Numbers", "Power"]]
Lottery_newdf.head()
```

```python
DoD=Lottery_newdf["Date"] 
WB1=Lottery_newdf["Winning Numbers"].str[0:2]
WB2=Lottery_newdf["Winning Numbers"].str[3:5]
WB3=Lottery_newdf["Winning Numbers"].str[6:8]
WB4=Lottery_newdf["Winning Numbers"].str[9:11]
WB5=Lottery_newdf["Winning Numbers"].str[12:14]
PBall=Lottery_newdf["Power"]
```

```python
Lottery_newdf2=pd.DataFrame({"White Ball 1": WB1,
                             "White Ball 2":  WB2,
                             "White Ball 3":  WB3,
                             "White Ball 4":  WB4,
                             "White Ball 5":  WB5,
                                "Power Ball": PBall}, dtype=np.int64)
Lottery_newdf2=Lottery_newdf2[["White Ball 1", "White Ball 2","White Ball 3","White Ball 4","White Ball 5", "Power Ball"]]
Lottery_newdf2.dropna(axis=0, how='all')
Indextot=len(Lottery_newdf2.index)
Indextot
```

```python
Lottery_newdf2.head()
```

```python
myNumbersIndex=np.arange(1, Indextot, dtype = int)
MyNumbersIndex=np.arange(1,Indextot-1, dtype =int)
```

```python
mynumberbins=myNumbersIndex
mynumberbins
```

```python
#Add up value counts on all White balls and create total table
newdf1=pd.cut(Lottery_newdf2["White Ball 1"], mynumberbins).value_counts()
newdf2=pd.cut(Lottery_newdf2["White Ball 2"], mynumberbins).value_counts()
newdf3=pd.cut(Lottery_newdf2["White Ball 3"], mynumberbins).value_counts()
newdf4=pd.cut(Lottery_newdf2["White Ball 4"], mynumberbins).value_counts()
newdf5=pd.cut(Lottery_newdf2["White Ball 5"], mynumberbins).value_counts()

newdf6=newdf1+newdf2+newdf3+newdf4+newdf5

newdf_dftotalWB = pd.DataFrame({"WB TotValCnts": newdf6, "WB TotValCnts%":newdf6/5/Indextot*100})

newdf_dftotalWB.set_index([MyNumbersIndex],inplace = True)
newdf_dftotalWB=newdf_dftotalWB.sort_values(by="WB TotValCnts", ascending=False)


newdf_dftotalWB=newdf_dftotalWB[0:68]

newdf_dftotalWB.head(15)
```

```python
newdfPB=pd.cut(Lottery_newdf2["Power Ball"], mynumberbins).value_counts()
newdf_dftotalPB = pd.DataFrame({"PB TotValCnts": newdfPB, "PB TotValCnts%": newdfPB/Indextot*100})
newdf_dftotalPB.set_index([MyNumbersIndex],inplace = True)
newdf_dftotalPB=newdf_dftotalPB.sort_values(by="PB TotValCnts", ascending=False)
newdf_dftotalPB=newdf_dftotalPB[0:24]
newdf_dftotalPB.head(10)
```
